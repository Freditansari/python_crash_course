```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np 
x = np.linspace(0,5,11)
y = x**2

potato_figure = plt.figure(figsize=(8,2), dpi = 100)
ax = potato_figure.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols = 1, figsize=(8,5))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('mypicture.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='potato distribution')
ax.plot(y,x,'R', label ='rice distribution')
ax.legend(loc = 0 )
# ax.legend(loc = (0.1,0.1) )


plt.tight_layout()
```
