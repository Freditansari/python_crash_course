```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('walmart_stock.csv', index_col ='Date', parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
df.rolling(7).mean().head(14)
```

```python
df.rolling(7).mean().plot()
```

```python
df['Open'].plot(figsize=(16,6))
df['Close'].rolling(window=30).mean().plot()
```

```python
df['Close 200 days MA'] = df['Close'].rolling(window=200).mean()
df[['Close 200 days MA', 'Close']].plot(figsize=(16,6))
```

```python
df['Close'].expanding().mean().plot(figsize=(16,6)) #expanding means is used to check if the price goes up or not
```

```python
#Bolinger bands
#Close 20 days MA
df['Close: 20 Days Means'] = df['Close'].rolling(20).mean()

# upper = 20 MA + 2*std(20)
df['Upper'] = df['Close: 20 Days Means'] + df['Close'].rolling(20).std()

#Lower = 20 MA - 2*std(20)
df['Lower'] = df['Close: 20 Days Means'] - df['Close'].rolling(20).std()

df[['Close', 'Close: 20 Days Means', 'Upper', 'Lower']].tail(20).plot(figsize=(45,16))
```
