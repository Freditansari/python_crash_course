```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

from mpl_finance import candlestick_ohlc
from matplotlib.dates import MONDAY, DateFormatter, DayLocator, WeekdayLocator


%matplotlib inline
```

```python
import pandas_datareader.data as web
import datetime
start = datetime.datetime(2017,1,1)
end = datetime.datetime(2019,1,23)
stocks = 'ERAA.JK MBSS.JK BMRI.JK'
stocks = stocks.split()
data = web.DataReader(stocks, 'yahoo', start, end )
```

```python
fig = plt.figure(figsize=(16,7))
ax = fig.add_axes([0,0,1,1])

for stock in stocks:
    ax.plot(data['Close', stock], label =stock)

ax.legend(loc = 0 )
```

```python
## transactions volume

volumesDF = pd.DataFrame()
fig = plt.figure(figsize=(16,7))
ax = fig.add_axes([0,0,1,1])

for stock in stocks:
    volumesDF[stock] = data['Open',stock] * data['Volume', stock]

for stock in stocks:
    ax.plot(volumesDF[stock], label =stock)

```

## Daily Returns & Volatility


```python
dailyReturnsDF = pd.DataFrame()

dailyReturnsDF['ERAA'] = data['Close','ERAA.JK'].pct_change(1) 
dailyReturnsDF['MBSS'] = data['Close','MBSS.JK'].pct_change(1) 
dailyReturnsDF['BMRI'] = data['Close','BMRI.JK'].pct_change(1)


```

```python
dailyReturnsDF['ERAA'].plot(kind='kde' , figsize = (20,8), label ='ERAA')
dailyReturnsDF['BMRI'].plot(kind='kde' , figsize = (20,8), label ='BMRI')
dailyReturnsDF['MBSS'].plot(kind='kde' , figsize = (20,8), label ='MBSS')

plt.legend()
```

```python
##this chart shows that PNBS is not anymore volatile than BBRI
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(dailyReturnsDF, figsize=(8,8), alpha = 0.2, hist_kwds={'bins':100})
```

## Cummulative returns

```python
Cummulative = pd.DataFrame()

Cummulative['ERAA'] = (1+dailyReturnsDF['ERAA']).cumprod() 
Cummulative['BMRI'] = (1+dailyReturnsDF['BMRI']).cumprod() 
Cummulative['MBSS'] = (1+dailyReturnsDF['MBSS']).cumprod()

```

```python
fig = plt.figure(figsize=(16,7))
plt.title('Cummulative Gains')

ax = fig.add_axes([0,0,1,1])

#can switch to for loop
ax.plot(Cummulative['ERAA'], label ='ERAA')
ax.plot(Cummulative['BMRI'], label ='BMRI')
ax.plot(Cummulative['MBSS'], label ='MBSS')

ax.legend(loc = 0 )
```
